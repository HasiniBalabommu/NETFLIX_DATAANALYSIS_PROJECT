```python
#we have to make the data in such way that Vote_average instead of numbers can be classified into categories such as 
#popluar average etcc and we can create a user defined function for that 
def categorize_col(data,col,labels):
    edges = [data[col].describe()['min'],
              data[col].describe()['25%'],
              data[col].describe()['50%'],
              data[col].describe()['75%'],
              data[col].describe()['max']]
    #pd.cut is a method in pandas which is used to categorize elements 
    data[col] = pd.cut(data[col],edges,labels = labels, duplicates = 'drop') # methods and col names 
    return data
```

```python
# like assigning the corresponding statements you want to print accross the data 
labels = ['not_popular','below_average','average','popular']

categorize_col(data,'Vote_Average',labels)
data['Vote_Average'].unique()
# the entire code snippet is commented because already execueted so not to throw error thats why 
```

```python
data['Genre'] = data['Genre'].str.split(',')
data = data.explode('Genre').reset_index(drop = True)
data['Genre'] = data['Genre'].str.strip()
data.head()
```
